Linear regression

```python

import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Sample data (hours studied, math score)
X = np.array([2, 4, 6, 8, 10]).reshape((-1, 1))
y = np.array([40, 60, 70, 80, 90])

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Predict math score for 7 hours studied
hours_studied = np.array([[7]])
predicted_score = model.predict(hours_studied)
print("Predicted math score:", predicted_score[0])

# Plot the data and regression line
plt.scatter(X, y)
plt.plot(X, model.predict(X), color="red")
plt.xlabel("Hours Studied")
plt.ylabel("Math Score")
plt.show()
```
